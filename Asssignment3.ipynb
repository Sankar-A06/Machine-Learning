#1.Basic Line Plot
Plot a simple line graph of ùë¶=ùë•^2 for ùë• values ranging from -10 to 10.

Label the x-axis as "X Values" and the y-axis as "Squared Values."

Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Define x values and calculate y values
x = np.linspace(-10, 10, 100)
y = x**2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2", color="blue")

# Label the axes and add a title
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Plot of y = x^2")

# Show the plot
plt.legend()
print("Here,It is a parabola")
print()
plt.show()
```

# 3.Scatter Plot
Generate two lists of 20 random integers each (for x and y coordinates).

Create a scatter plot with these values, and customize it by adding color,size, and labels to the points.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate two lists of 20 random integers each for x and y coordinates
np.random.seed(42)  # For reproducibility
x = np.random.randint(1, 101, 20)
y = np.random.randint(1, 101, 20)

# Generate random colors and sizes for the points
colors = np.random.rand(20)
sizes = np.random.randint(50, 300, 20)

# Create a scatter plot
plt.figure(figsize=(8, 6))
z= plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap='viridis', edgecolor='black')

# Add labels and title
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.title("Customized Scatter Plot")

# Show color bar for the colors
plt.colorbar(z, label="Color Scale")

# Show the plot
plt.grid(True)
plt.show()
```

# 6.Box Plot (Seaborn)
Create a box plot to visualize the distribution of the marks scored by students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]. o Highlight the median and any outliers.

```python

marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
sns.boxplot(data=marks)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Box Plot of Marks Scored by Students in a class')
# Display the plot
plt.show()
```

Highlight the median and any outliers

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

sns.boxplot(data=marks)

median = np.median(marks)
plt.axhline(median, color='g', linestyle='--', label='Median')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Box Plot of Marks Scored by Students')

plt.legend()
plt.show()
```

# 8.
Comparison of Matplotlib and Seaborn o Generate a pair of plots (barcharts or line plots) using the same data in both Matplotlib and Seaborn Highlight the differences in the code, syntax, and aesthetics of the two
libraries.


```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Matplotlib')
# Display the plot
plt.show()
```

#Difference in code , Aesthitics and syntax

#Code and Syntax:
Matplotlib: The code is more straightforward and requires fewer imports. You directly use the plt.bar() function to create the bar chart.

**Seaborn**: You need to create a DataFrame to use Seaborn effectively. The sns.barplot() function is used, and you pass the DataFrame along with the column names for the x and y axes.

#Aesthetics:
***Matplotlib***: The default aesthetics are simpler and more basic. Customization requires more manual adjustments.

***Seaborn***: The default aesthetics are more polished and visually appealing. Seaborn automatically applies a more sophisticated style to the plots.

# 10.	Save and Style Plots

‚Ä¢	Plot any graph of your choice and save it in both .png and .pdf formats.

‚Ä¢	Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid')

and explain how the style impacts the visualization.


```python
import matplotlib.pyplot as plt
import numpy as np

# Apply the 'ggplot' style
plt.style.use('ggplot')

# Generate x values
x = np.linspace(0, 10, 100)

# Compute y values
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='sin(x)')

# Customize the plot
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Plot of sin(x)')
plt.legend()

# Save the plot in .png and .pdf formats
plt.savefig('plot.png')
plt.savefig('plot.pdf')

# Display the plot
plt.show()
```

### The **'ggplot'** style is inspired by the aesthetics of the ggplot2 package in R.

It provides a clean and modern look to the plots with grid lines, a light gray background, polished font and line styles, and a vibrant color palette.
